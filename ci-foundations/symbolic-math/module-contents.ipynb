### Introduction to Symbolic Algebra in Python

The purpose of this lesson is to help students develop foundational skills for performing symbolic algebra in Python using the `sympy` and `algebra_with_sympy` libraries (below, referred to collectively as SymPy), with some brief applications to problem solving at the general chemistry (first-year) level.

### Lesson Learning Outcomes (LOs)
At the end of this lesson, students will be able to...

#### Cyberinfrastructure LOs
1. Define mathematical variables and expressions using SymPy
2. Manipulate algebraic expressions to solve for a single variable
3. Substitute numerical values and units into algebraic expressions
4. Apply symbolic algebra techniques in Python to solve general chemistry problems

### Introduction to Symbolic Calculus in Python

The purpose of this lesson is to help students develop skills in performing symbolic calculus manipulations of scalar functions of single or multiple variables using SymPy.

### Lesson Learning Outcomes (LOs)
At the end of this lesson, students will be able to...

#### Cyberinfrastructure LOs

1. Evaluate limits of scalar functions of one or multiple variables
2. Perform symbolic derivatives of scalar functions of one or multiple variables
3. Perform indefinite integrals of scalar functions of one or multiple variables
4. Evaluate definite integrals of scalar functions of one or multiple variables
5. Visualize SymPy expressions graphically

### Lesson Prerequisites
#### CI Prereqs
- [Basic Python Syntax](../python-basics/python-syntax.ipynb)
- [Symbolic Algebra](symbolic-algebra.ipynb)

#### Content Prereqs
This lesson is focused on the development of the _cyberinfrastructure skills_ for performing symbolic calculus operations on real-valued scalar functions in Python rather than teaching the calculus itself. Students are therefore expected to be able to perform by hand the following operations from single-variable calculus:
- Limits of scalar functions of a single variable
- Ordinary derivatives of scalar functions of a single variable
- Indefinite integrals of scalar functions of a single variable
- Definite integrals of scalar functions of a single varialbe

Because this lesson is developed with a student in a science course as the intended audience, however, some brief conceptual reminders of the behavior of these operations are provided to bring students back up to speed.

~~~{note} Note: Multivariate Calculus
No familiarity with the calculus of scalar functions of multiple variables is assumed. Instead, these concepts are introduced as natural extensions of the calculus of scalar functions of a single variable.
~~~

### Symbolic Representation of Operators in SymPy

The purpose of this lesson is to help students develop the skill of symbolically applying mathematical operators to scalar functions using SymPy. 

### Lesson Learning Outcomes (LOs)
At the end of this lesson, students will be able to...

#### Cyberinfrastructure LOs
1. Construct the SymPy representation of an operator from its mathematical definition,
2. Apply linear and nonlinear operators to definite SymPy expressions,
3. Apply linear and nonlinear operators to arbitrary SymPy functions,
4. Evaluate relationships of mathematical equality and inequality between SymPy expressions
